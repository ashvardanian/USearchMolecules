```python
import os
import pandas as pd

# Specify the directory path
directory = 'data/pubchem/parquet/'

# Get a list of all parquet files in the directory
parquet_files = [file for file in os.listdir(directory) if file.endswith('.parquet')]

# Sort the list of parquet files alphabetically
parquet_files.sort()

# Read the first parquet file into a dataframe
df = pd.read_parquet(os.path.join(directory, parquet_files[0]))

# Display the dataframe
df.head()
```
